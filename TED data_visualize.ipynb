```python
#Buổi nói chuyện mà tạo ra nhiều cuộc thảo luận nhất:
#Ý tưởng: dựa vào cột comments để xem, số cmt cao nhất sẽ là kết quả? thử nào:
ted.sort_values('comments', ascending = False).head()
```

```python
#Tuy nhiên có gì đó sai sai ở đây:
# các buổi talk có thời lượng không giống nhau, có talk ngắn có talk dài, như vậy là không ổn nếu như chỉ dựa vào cột comments.
# => tạo một cột mới: số bình luận trên mỗi lượt xem, số lượt xem trên mỗi bình luận :D
ted['cmt/view'] = ted['comments'] / ted['views']
ted.sort_values('cmt/view', ascending = False).head()
```

```python
ted['view/cmt'] = ted['views'] / ted['comments']
ted.sort_values('view/cmt', ascending = True).head()
```
